```python

# Binomial Distribution (n=10, p=0.5)
import numpy as np
samples = np.random.binomial(n=10, p=0.5, size=1000)

import matplotlib.pyplot as plt
plt.hist(samples, bins=range(12), edgecolor='black', alpha=0.75)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Normal Distribution (μ=0, σ=1)

samples = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(samples, bins=30, edgecolor='black', alpha=0.75)
plt.title('Normal Distribution (μ=0, σ=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Uniform Distribution (0, 1)

samples = np.random.uniform(low=0, high=1, size=1000)
plt.hist(samples, bins=30, edgecolor='black', alpha=0.75)
plt.title('Uniform Distribution (0, 1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
