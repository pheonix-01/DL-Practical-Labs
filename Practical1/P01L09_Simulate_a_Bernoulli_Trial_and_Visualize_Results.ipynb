```python
# Simulate a Bernoulli Trial and Visualize Results
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p = 0.3  # Probability of heads (1)
n = 1000  # Number of trials

# Simulate Bernoulli trials (0 = tails, 1 = heads)
trials = np.random.binomial(n=1, p=p, size=n)

# Count occurrences
unique, counts = np.unique(trials, return_counts=True)
results = dict(zip(unique, counts))

# Map 0 → "Tails", 1 → "Heads"
labels = ['Tails', 'Heads']
frequencies = [results.get(0, 0), results.get(1, 0)]

# Plot
plt.bar(labels, frequencies, color=['skyblue', 'orange'])
plt.title(f"Bernoulli Trials (n = {n}, p = {p})")
plt.ylabel("Frequency")
plt.xlabel("Outcome")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
