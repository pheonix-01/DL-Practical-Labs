```python
# Gradient Descent on a Multivariable Function
import numpy as np
import matplotlib.pyplot as plt

# Function and its gradient
def f(x, y):
    return x**2 + y**2

def grad_f(x, y):
    return np.array([2*x, 2*y])  # Gradient vector [df/dx, df/dy]

# Parameters
x, y = 2.0, 2.0          # Initial values
alpha = 0.1              # Learning rate
iterations = 100         # Number of steps

# Lists to track the path
x_vals = [x]
y_vals = [y]
f_vals = [f(x, y)]

# Gradient descent loop
for _ in range(iterations):
    grad = grad_f(x, y)
    x -= alpha * grad[0]
    y -= alpha * grad[1]
    x_vals.append(x)
    y_vals.append(y)
    f_vals.append(f(x, y))

# Prepare contour plot of f(x, y)
X, Y = np.meshgrid(np.linspace(-2.5, 2.5, 100), np.linspace(-2.5, 2.5, 100))
Z = f(X, Y)

# Plotting the contour and path
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=30, cmap='viridis')
plt.clabel(contour, inline=1, fontsize=8)
plt.plot(x_vals, y_vals, marker='o', color='red', label='Gradient Descent Path')
plt.scatter([x_vals[0]], [y_vals[0]], color='blue', label='Start')
plt.scatter([x_vals[-1]], [y_vals[-1]], color='green', label='End (Minimum)')
plt.title("Gradient Descent Path on f(x, y) = x + y")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
