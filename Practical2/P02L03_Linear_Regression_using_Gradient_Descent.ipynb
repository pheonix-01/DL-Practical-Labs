```python
# Linear Regression using Gradient Descent
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3x + noise

# Add a bias term x = 1 for 
X_b = np.c_[np.ones((100, 1)), X]  # shape: (100, 2)

# Step 2: Initialize parameters
theta = np.random.randn(2, 1)  # [, ]
learning_rate = 0.1
iterations = 100
m = len(X_b)  # number of samples

cost_history = []

# Step 3: Gradient Descent loop
for i in range(iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients
    cost = np.mean((X_b.dot(theta) - y) ** 2)
    cost_history.append(cost)

# Step 4a: Plot cost vs iterations
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(iterations), cost_history, color='red')
plt.title("Cost vs Iterations")
plt.xlabel("Iteration")
plt.ylabel("Cost (MSE)")
plt.grid(True)

# Step 4b: Plot data and fitted line
plt.subplot(1, 2, 2)
plt.scatter(X, y, color='blue', label='Data')
x_line = np.array([[0], [2]])
x_line_b = np.c_[np.ones((2, 1)), x_line]
y_line = x_line_b.dot(theta)
plt.plot(x_line, y_line, color='green', label='Fitted Line')
plt.title("Linear Regression Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Print final theta values
print(f"Final : {theta[0][0]:.4f}, : {theta[1][0]:.4f}")
```
