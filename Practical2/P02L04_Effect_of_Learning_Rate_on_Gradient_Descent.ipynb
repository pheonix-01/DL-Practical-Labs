```python
# Effect of Learning Rate on Gradient Descent
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) and its derivative f'(x)
def f(x):
    return x**4 - 3*x**3 + 2

def df(x):
    return 4*x**3 - 9*x**2

# Gradient descent implementation
def gradient_descent(start_x, learning_rate, iterations):
    x_vals = [start_x]
    x = start_x
    for _ in range(iterations):
        grad = df(x)
        x -= learning_rate * grad
        x_vals.append(x)
    return x_vals

# Parameters
start_x = 0.5
iterations = 50
learning_rates = [0.001, 0.01, 0.1]
colors = ['red', 'green', 'blue']

# Plot the function
x_range = np.linspace(-1, 3, 400)
y_range = f(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label="f(x) = x⁴ - 3x³ + 2", color='black')

# Run gradient descent for each learning rate
for alpha, color in zip(learning_rates, colors):
    x_path = gradient_descent(start_x, alpha, iterations)
    y_path = [f(x) for x in x_path]
    plt.plot(x_path, y_path, marker='o', color=color, label=f'α = {alpha}')

# Plot settings
plt.title("Gradient Descent Paths for Different Learning Rates")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
